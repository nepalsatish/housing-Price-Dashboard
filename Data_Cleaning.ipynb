```python

import pandas as pd
```

```python
df = pd.read_csv('/var/www/project/Housing/zhvi_data.csv')
```

```python
df.head()
```

```python
id_cols = ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName']
df_long = pd.melt(df, id_vars=id_cols, var_name='Date', value_name='Price')
df_long['Date'] = pd.to_datetime(df_long['Date'])
df_long = df_long.dropna(subset=['Price'])
df_long.head()
```

```python
# Remove state initials from RegionName column
df_long['RegionName'] = df_long['RegionName'].str.split(',').str[0]

# Save the cleaned data
df_long.to_csv('/var/www/project/Housing/zhvi_data_long.csv', index=False)
print("Data cleaned and saved successfully!")
print(f"Shape of cleaned data: {df_long.shape}")
```

```python
test_cities = ['New York', 'Los Angeles', 'Chicago', 'Dallas', 'San Francisco']
df_test = df_long[df_long['RegionName'].isin(test_cities)]
df_test.to_csv('/var/www/project/Housing/zhvi_data_test.csv', index=False)
df_test.head()
```
